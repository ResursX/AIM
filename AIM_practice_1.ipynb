```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Lab1\\data\\titanic.csv')
```

```python
df
```

### 1. Разбиение на таблицы по полу

C удалением пустых строк

```python
df = df.dropna()

df_male = df[df['Sex'] == 'male']
df_female = df[df['Sex'] == 'female']
```

```python
df_male
```

```python
df_female
```

### 2. Гистограмма распределения пассажиров по возрасту

```python
df['Age'].plot(kind='hist', bins=50)
plt.xlabel('Возраст')
plt.ylabel('Кол-во пассажиров')
plt.grid(axis='y')
plt.show()
```

### 3. Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола

```python
df.groupby('Sex').agg({'Age': ['min', 'max', 'mean', pd.Series.mode, 'median']}, axis='columns')
```

### 4. Визуализировать возраст при помощи различных настроек метода plot

```python
df_male['Age'].plot(kind='line', label='Мужчины', color='blue', alpha=0.6)
df_female['Age'].plot(kind='line', label='Женщины', color='purple', alpha=0.6)

plt.title('Линейный график возраста')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```

```python
df_male['Age'].plot(kind='hist', label='Мужчины', color='blue', alpha=0.7, bins=15)
df_female['Age'].plot(kind='hist', label='Женщины', color='purple', alpha=0.7, bins=15)

plt.title('Гистограмма возраста')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

```python
#df_male['Age'].plot(kind='box', x=1, label='Мужчины', color='blue')
#df_female['Age'].plot(kind='box', x=2, label='Женщины', color='purple',)

df.plot.box(column='Age', by='Sex')

plt.title('Коробчатая диаграмма')
plt.ylabel('Возраст')
plt.show()
```

```python
df_male['Age'].plot(kind='kde', label='Мужчины', color='blue', alpha=0.6)
df_female['Age'].plot(kind='kde', label='Женщины', color='purple', alpha=0.6)

plt.title('Диаграмма плотности')
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

```python
plt.scatter(df_male.index, df_male['Age'], label='Мужчины', color='blue', alpha=0.6)
plt.scatter(df_female.index, df_female['Age'], label='Женщины', color='purple', alpha=0.6)

plt.title('Точечная диаграмма')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```
