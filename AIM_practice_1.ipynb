```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Lab1\\data\\titanic.csv')
```

```python
df
```

### 1. Разбиение таблицы по полу

C удалением пустых строк

```python
df = df.dropna()

df_male = df[df['Sex'] == 'male']
df_female = df[df['Sex'] == 'female']
```

```python
df_male
```

```python
df_female
```

### 2. Гистограмма распределения пассажиров по возрасту

```python
df['Age'].plot(kind='hist', bins=50)
plt.xlabel('Возраст')
plt.ylabel('Кол-во пассажиров')
plt.grid(axis='y')
plt.show()
```

### 3. Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола

```python
df.groupby('Sex').agg({'Age': ['min', 'max', 'mean', pd.Series.mode, 'median']}, axis='columns')
```

### 4. Визуализировать возраст при помощи различных настроек метода plot

```python
df_male['Age'].plot(kind='line', label='Male', color='blue')
df_female['Age'].plot(kind='line', label='Female', color='purple')

plt.title('Age lineplot')
plt.ylabel('Age')
plt.show()
```

```python
df_male['Age'].plot(kind='hist', label='Male', color='blue', alpha=0.7, bins=15)
df_female['Age'].plot(kind='hist', label='Female', color='purple', alpha=0.7, bins=15)

plt.title('Age histogram')
plt.ylabel('Age')
plt.show()
```

```python
df_male['Age'].plot(kind='box', label='Male', color='blue')
df_female['Age'].plot(kind='box', label='Female', color='purple')

#plt.title('Age boxplot')
#plt.ylabel('Age')
#plt.show()
```

```python
plt.scatter(df_male.index, df_male['Age'], label='Male', color='blue', alpha=0.6)
plt.scatter(df_female.index, df_female['Age'], label='Female', color='purple', alpha=0.6)

plt.title('Точечная диаграмма')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```
